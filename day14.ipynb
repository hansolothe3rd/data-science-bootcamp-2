Focus: KDE, heatmaps, scatterplots with regression, and subplot layouts.


**Goals for Day 14**
- Practice Kernel Density Estimation (KDE) plots
- Explore correlation heatmaps
- Build scatterplots with regression lines
- Combine multiple plots into subplots for comparison


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("titanic_clean.csv")
```

1. KDE Plot

```python
sns.kdeplot(df['age'], fill=True, color='blue')
plt.title("Age Distribution (KDE)")
plt.xlabel("Age")
plt.show()
```

2. Correlation Heatmap

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

3. Scatterplot with Regression

```python
sns.lmplot(data=df, x="age", y="fare", aspect=1.5, line_kws={'color':'red'})
plt.title("Age vs Fare with Regression Line")
plt.show()
```

4. Subplots Example

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.boxplot(ax=axes[0], x="sex", y="age", data=df)
axes[0].set_title("Age Distribution by Sex")

sns.violinplot(ax=axes[1], x="sex", y="fare", data=df)
axes[1].set_title("Fare Distribution by Sex")

plt.tight_layout()
plt.show()
```

5. Reflection

### Reflection
- KDE plots are great for smooth distributions
- Heatmaps reveal correlations between numeric variables
- Regression plots show trends between variables
- Subplots allow side-by-side comparisons for storytelling
