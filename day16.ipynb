**Goals for Day 16**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Descriptive Statistics

```python
# Using Wine dataset from Day 15
wine = pd.read_csv("wine_clean.csv")

# Descriptive stats
wine.describe()

# Example: mean, median, std of alcohol
print("Mean alcohol:", wine['alcohol'].mean())
print("Median alcohol:", wine['alcohol'].median())
print("Std alcohol:", wine['alcohol'].std())
```

2. Probability Distributions

```python
# Normal distribution example
mu, sigma = 0, 1
normal_samples = np.random.normal(mu, sigma, 1000)

plt.figure(figsize=(8,5))
sns.histplot(normal_samples, bins=30, kde=True)
plt.title("Normal Distribution (mean=0, std=1)")
plt.show()

# Uniform distribution example
uniform_samples = np.random.uniform(0, 10, 1000)

plt.figure(figsize=(8,5))
sns.histplot(uniform_samples, bins=30, kde=True)
plt.title("Uniform Distribution (0-10)")
plt.show()
```

3. Central Limit Theorem Demo

```python
# Take samples from wine alcohol column
sample_means = []
for _ in range(1000):
    sample = wine['alcohol'].sample(50, replace=True)
    sample_means.append(sample.mean())

plt.figure(figsize=(8,5))
sns.histplot(sample_means, bins=30, kde=True)
plt.title("Central Limit Theorem: Sample Means of Alcohol")
plt.show()

print("Mean of sample means:", np.mean(sample_means))
print("Std of sample means:", np.std(sample_means))
```

4. Reflection

### Reflection
