# ðŸ“˜ Day 19: Correlation & Covariance

## Correlation and covariance are tools to measure the relationship between variables in your dataset.

## ðŸ§  Key Concepts

- Covariance

    - Measures how two variables change together.

    - Positive â†’ variables increase together.

    - Negative â†’ one increases while the other decreases.

    - Scale-dependent â†’ hard to compare across variables.

- Correlation

    - Standardized version of covariance.

    - Always between -1 and 1.

    - Easier to interpret:

        - 1 â†’ perfect positive relationship

        - -1 â†’ perfect negative relationship

        - 0 â†’ no linear relationship

- Use case in data science:

    - Identify relationships between features and target variables.

    - Spot multicollinearity (when predictors are too correlated).

## ðŸ“Š Example with Wine Dataset

Weâ€™ll look at correlations between chemical properties and wine quality.

### ðŸ”¹ Load Data

```python
import pandas as pd

# Load cleaned dataset
wine = pd.read_csv("wine_clean.csv")
wine.head()
```

### ðŸ”¹ Covariance Matrix

```python
# Covariance shows raw relationships
wine.cov()
```

### ðŸ”¹ Correlation Matrix

```python
# Correlation standardizes the covariance
wine.corr()
```

### ðŸ”¹ Heatmap for Better Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(wine.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Wine Dataset")
plt.show()
```

### âœ… What to Look For

- Which variables are most correlated with quality?

- Any strong correlations between predictors (multicollinearity risks)?

### ðŸ”¹ Correlation Matrix

```python
# Compute correlation matrix
corr_matrix = wine.corr()

# Look at correlations with quality
quality_corr = corr_matrix["quality"].sort_values(ascending=False)
quality_corr
```

### ðŸ”¹ Top Correlations with quality

```python
# Display top positive & negative correlations with quality
print("Top correlations with wine quality:\n")
print(quality_corr.head(5))   # top positive
print("\n")
print(quality_corr.tail(5))   # top negative
```

## âœ… Insights to Watch For

- alcohol usually has a positive correlation with quality.

- volatile_acidity often has a negative correlation with quality.

- Weak correlations (< |0.1|) may not be useful predictors.
